## Identify skewness and outliers

```python
skew_value = df['Age'].skew()
print(f'Skewness of Age column: {skew_value}')

if skew_value > 0:
  print('Right skewed (positive skew): tail is longer on the right')
elif skew_value < 0:
  print('Left skewed (negative skew): tail is longer on the left')
else:
  print('Symmetrical (normal distribution)')
```

```python
#Assuming df is your DataFrame and 'Age' is the column of interest
Q1 = df ['Age'].quantile(0.25)
Q3 = df ['Age'].quantile(0.75)
IQR = Q3 - Q1

# Define bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
#Filter out the outliers
outliers = df[(df['Age'] < lower_bound) | (df['Age'] > upper_bound)]

print(outliers)
```

```python
#Assuming df is your DataFrame and 'Weight' is the column of interest
Q1 = df ['Weight'].quantile(0.25)
Q3 = df ['Weight'].quantile(0.75)
IQR = Q3 - Q1

# Define bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
#Filter out the outliers
outliers = df[(df['Weight'] < lower_bound) | (df['Weight'] > upper_bound)]

print(outliers['Weight'].unique())
```

## Create a timedata column and categories into groups

```python
df['Time_data'] = df['Treatment_Start_Date'] - df['Diagnosis_Date']

df['Time_data']
```

```python
df['Time_data'].describe()
```

```python
#check for the number of unique Time_data
print('Count of unique Time_data :', df['Time_data'].nunique())

#Create bins for the various age categories
min_day = df['Time_data'].min()
twenty_fifth = df['Time_data'].quantile(0.25)
median = df['Time_data'].median()
seventy_fifth = df['Time_data'].quantile(0.75)
maximum =df['Time_data'].max()

#Create a Label and bins for the various age categories
labels = ['Immediate Start', 'Early Start', 'Moderate Delay', 'Delayed Start']
bins = [min_day, twenty_fifth, median, seventy_fifth, maximum]

#Create a category column for age
df['Time_data_cat']=pd.cut(df['Time_data'], labels=labels, bins=bins)
```
